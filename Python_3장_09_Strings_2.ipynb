# 문자열의 메쏘드는 아주 많다.

```
a = 'Hello Python'

print(a.count('o')) # 파라미터 값의 존재 횟수 반환
print(a.find('t')) # 파라미터 값이 있는 위치 반환
print(a.index('t')) # 파라미터 값이 있는 위치 반환
```

```
# 그럼, find() 함수와 index() 함수의 차이는 ?
# find()는 없을때는 -1을 반환
print(a.find('a'))
```

```
# index()는 없을때는 error, 즉 index() 함수는 파라미터 값이 문자열에 존재한다는 것을 가정하고 있다.

print(a.index('a'))
```

```
# 따라서 문자열의 index() 메쏘드를 사용하려면 먼저 in 으로 검사하고 사용하자.
a = 'Hello Python'

if 'a' in a:
    print(a.index('t'))
```

```
# 유용하게 사용되는 몇개의 메쏘드를 더 봅시다.

print('python'.upper()) # upper()은 대문자로 변경된 새로운 문자열을 "반환"
print('Python'.lower()) # lower()은 소문자로 변경된 새로운 문자열을 "반환"
```

```
# 다시 말하지만, 문자열 그 자체를 변경하는 것이 아니다.
# 변경된 새로운 문자열을 "반환"한다.

hello_msg = "hello python"
print(hello_msg.upper())
print(hello_msg)
```

```
# 문자열 내부의 각 단어의 첫글자만 대문자로 변경할 수 있을까요?

hello_msg = "hello python, my name is keechul jung."
output_msg = ""
need_to_change = False

for i, ch in enumerate(hello_msg):
    if i == 0:
        output_msg += ch.upper()
    elif ch == " ":
        output_msg += ch
        need_to_change = True
    else:
        if need_to_change == True:
            output_msg += ch.upper()
            need_to_change = False
        else:
            output_msg += ch

print(output_msg)
```

```
# 그렇지만, 파이썬은 많은 문자열 함수 관련 기능이 이미 제공됩니다.
# 위와 동일한 결과를 1줄로 구현할 수 있다.

hello_msg = "hello python, my name is keechul jung."
print(hello_msg.capitalize()) # 문자열의 첫 글자만 대문자로 변경한 새로운 문자열을 반환
print(hello_msg.title()) # 문자열의 모든 단어의 첫 글자를 대문자로 변경한 새로운 문자열을 반환
print(hello_msg)
```

# 빈칸 정리와 관련된 메쏘드를 살펴보자.

```
# 공백을 정리

print("   -python-   ".lstrip()) # left strip
print("   -python-   ".rstrip()) # right strip
print("   -python-   ".strip())  # strip
```

```
# 아래와 같이 "단어들 사이 사이의 공백"은 어떻게 정리할 수 있을까?
# 앞에서 배운 strip() 계열의 메쏘드로는 않된다.

print("   -i  love        python-      ".lstrip())
print("   -i      love  python-      ".rstrip()) 
print("   -i      love     python-      ".strip())
```

```
# replace() 메쏘드를 살펴보자.

print("I love python".replace("I", "You"))
```

```
# 공백 2개를 1개로 바꾸면... 될 것 같아서
# 아래와 같이 replace() 메쏘드를 쓰면 될까? 했었는데...
# 않된다.

print("    'i    love  python'   ".replace("    ", " "))
```

# 파이썬에서 아주 많이 사용하는 join()과 split() 메쏘드를 사용해보자.
 - 리스트 자료형을 알아야해서, 간단하게 리스트 자료형 설명도 참고하자.
 - 리스트 자료형의 자세한 내용은 후에 설명된다.


```
# 문자열의 split() 메쏘드는 해당하는 파라미터를 기준으로 
# 문자열을 단어별로 구분해서 
# 리스트로 반환한다.

print("I love python".split())
print("I_love_python. You will do too".split('_'))
```

```
# 리스트 자료형 설명 간단하게하자.
# 파이썬의 리스트는 혹시 기존 언어를 배웠으면 '배열'이라고 생각하면 된다.
# 다른 언어를 배우지 않았으면, 여러 개의 값을 인덱스로 관리하는 자료형이라고 생각하자.

number_list = [1,2,3,4,5]
print(number_list[1])

class_list = ['Math', 'English', 'Science']
print(class_list[1])
```

```
# 문자열 자료형의 split() 메쏘드는 아래와 같이 사용할 수 있다.

word_list = "I_love_python".split('_')

print(word_list)

print(word_list[0])
print(word_list[1])
print(word_list[2])

for w in word_list:
    print(w)
```

```
# 위에서 리스트로 분할된 단어들을 
# 문자열의 join() 메쏘드를 이용해서 다시 문자열로 만들수 있다.
# 즉, 문자열의 join() 메쏘드는 파라미터로 리스트를 받아서 이를 연결한 문자열을 반환한다.
```

```
print(" ".join(word_list))
print(type(" ".join(word_list)))
```

```
# 즉, join 메쏘드는 아래와 같이 동작한다.

print(" ".join(['python','programming','too','interesting']))
print("*".join(['python','programming','too','interesting']))
print("___".join(['python','programming','too','interesting']))
```

# 총 정리하면...
- 우리는 지금 챗봇의 입력 문장에 불필요한 공백을 정리하려고 한다.
- 아래와 같이 정리하면 문자열의 공백들을 정리할 수 있다.

```
msg = " I    Love Python    Very    Much    "
msg_list = msg.split()
msg = " ".join(msg_list)
msg = msg.lower()
print(msg)
```

```
# 위의 코드를 아래와 같이 간소화할 수 있다.
# 복잡해 보이지만, 잘 뜯어서 보자.

msg = " I    Love Python    Very    Much    "
# 아래의 문장을 이해하기 위해서는 split() 또는 lower() 등의 문자열 메쏘드의 반환값이 문자열이라는 것을 확실히 이해해야한다.
msg = " ".join(msg.split()).lower()
print(msg)
```
