## 일차함수 Gradient Descent 구현하기

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(10)
x = np.random.rand(20).round(2)
y = 0.7 * x
print(x)
print(y)
```

```python
plt.figure(figsize = (4,3))
plt.scatter(x,y)
plt.show()
```

```python
W = np.random.uniform(0,1)
b = np.random.uniform(0,1)
y_pred = W * x + b
def compared_pred(x, pred, y):
    plt.figure(figsize=(4,3))
    plt.scatter(x,y, label = 'ground-truth')
    plt.scatter(x, pred, label = 'prediction')
    plt.legend()
    plt.show()
pred = W*x +b
compared_pred(x,pred,y)
```

```python
W = np.random.uniform(0,1)
b = np.random.uniform(0,1)

step_size = 0.5

for epoch in range(100):
    y_pred = W* x + b

    #mean,sum 상관없음
    objective = np.abs(y_pred - y).mean()
    
    #위 함수를 w,b에 대해서 미분 
    #원래 식에 따라 y_pred - p 에 x를 곱한것 만큼이 w_grad
    w_grad = step_size*((y_pred-y)*x).mean()
    b_grad = step_size*((y_pred-y)).mean()

    W = W- w_grad
    b = b- b_grad
    if epoch%10 == 9:
        print('epoch:', epoch+1,',objective:', objective)
        print('W:',W, ',b:',b)
```

```python
pred = W*x +b
compared_pred(x,pred,y)
```
